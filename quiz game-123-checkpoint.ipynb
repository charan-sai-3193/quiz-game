```python


quiz = [
    {
        "question": "What is the capital of France?",
        "options": ["A) London", "B) Berlin", "C) Paris", "D) Madrid"],
        "answer": "C"
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "options": ["A) Venus", "B) Mars", "C) Jupiter", "D) Saturn"],
        "answer": "B"
    },
    {
        "question": "What is 9 x 8?",
        "options": ["A) 72", "B) 81", "C) 64", "D) 69"],
        "answer": "A"
    },
    {
        "question": "Who painted the Mona Lisa?",
        "options": ["A) Van Gogh", "B) Da Vinci", "C) Picasso", "D) Michelangelo"],
        "answer": "B"
    },
    {
        "question": "Which gas do plants absorb from the atmosphere?",
        "options": ["A) Oxygen", "B) Nitrogen", "C) Carbon Dioxide", "D) Helium"],
        "answer": "C"
    },
    {
        "question": "What is the largest mammal?",
        "options": ["A) Elephant", "B) Blue Whale", "C) Giraffe", "D) Orca"],
        "answer": "B"
    },
    {
        "question": "Who wrote the play 'Romeo and Juliet'?",
        "options": ["A) Oscar Wilde", "B) Jane Austen", "C) Shakespeare", "D) Mark Twain"],
        "answer": "C"
    },
    {
        "question": "What is the boiling point of water at sea level?",
        "options": ["A) 100°C", "B) 90°C", "C) 80°C", "D) 110°C"],
        "answer": "A"
    },
    {
        "question": "Which language is the most widely spoken worldwide?",
        "options": ["A) English", "B) Hindi", "C) Mandarin Chinese", "D) Spanish"],
        "answer": "C"
    },
    {
        "question": "In computing, what does 'CPU' stand for?",
        "options": ["A) Central Processing Unit", "B) Computer Power Unit", "C) Control Panel Utility", "D) Core Program Unit"],
        "answer": "A"
    }
]

def run_quiz(quiz):
    score = 0
    for i, q in enumerate(quiz):
        print(f"\nQuestion {i+1}: {q['question']}")
        for option in q['options']:
            print(option)
        
        while True:
            answer = input("Your answer (A/B/C/D): ").strip().upper()
            if answer in ['A', 'B', 'C', 'D']:
                break
            else:
                print("Please enter a valid option (A, B, C, or D).")
        
        if answer == q["answer"]:
            print("Correct! ✅")
            score += 1
        else:
            print(f"Wrong ❌. The correct answer was {q['answer']}.")
    
    print(f"\nQuiz Completed! You got {score} out of {len(quiz)} correct.")


run_quiz(quiz)
```
